```python
# 1. Difference between "Love" and "love" in NLP

# In Natural Language Processing (NLP), the words "Love" and "love" are treated as different tokens if text normalization is not applied. The word "Love" is capitalized and may indicate the beginning of a sentence or emphasis, while "love" is the standard lowercase form. If lowercasing is not performed, it increases the vocabulary size unnecessarily, making the model less efficient.

# 2. What happens if stopwords are not removed?

# If stopwords are not removed, they add unnecessary noise to the data. This makes the model unnecessarily complex, which in turn affects its accuracy. The accuracy is compromised because important words in the sentence get overshadowed by stopwords.

# 3. When removing stopwords can be harmful
# Removing stopwords is not always good for the model. Sometimes, it changes the meaning of the sentence. 

# Sentiment Analysis:
# For example, in the sentence "I am not happy," removing stopwords changes the entire meaning of the sentence. 
# Question Answering / Chatbots:
# Similarly, in the sentence "What is your name?" removing stopwords changes the entire meaning.

# 4. Difference between Stemming and Lemmatization

# Stemming and Lemmatization are two techniques used in the field of NLP that convert words into their root form, but the difference between the two techniques is that the stemming technique uses a rule-based approach that simply removes the prefixes or suffixes of the words, resulting in a non-meaningful word in some cases (for example, the word "running" would be reduced to "runn"). However, the lemmatization technique uses a dictionary-based approach that considers the context in which the word has been used in the text to convert it into a meaningful root word, also known as a lemma (for example, the word "running" would be reduced to "run").
```

```python
#TASK 2
import re

def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs & emails
    text = re.sub(r'http\S+|www\S+|@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (soooo → soo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters (keep words)
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    # Remove short words (≤2) except 'no', 'not'
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence
```

```python
#TASK 3

test_sentences = [
    "Get 100% FREE access now!",
    "I absolutely looooved this product",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    results.append((sentence, tokens, clean))

for r in results:
    print("\nOriginal:", r[0])
    print("Tokens:", r[1])
    print("Clean Sentence:", r[2])
```

```python
import numpy as np

for original, tokens, clean in results:
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(t) for t in tokens]) if tokens else 0

    print("\nSentence:", original)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Avg Token Length:", round(avg_length, 2))
```

```python
from collections import Counter

all_tokens = []
for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:", counter.most_common(10))
print("Least 5 Frequent Words:", counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
