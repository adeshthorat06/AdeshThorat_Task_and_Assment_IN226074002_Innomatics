```python
#Problem statement: 1
class control_smartlight:
    def __init__(self, light_name): #Constructor to initialize
        self.light_name = light_name
        self.status = "OFF"
    
    def turn_on(self):
        self.status = "ON"
    def turn_off(self):
        self.status = "OFF"
    def display_status(self):
       
        print(f"{self.light_name} is {self.status}")

light = control_smartlight("Bedroom Light")
light.turn_on()
light.display_status()

        
```

```python
#Problem statement No: 2

class Employee_ID:
    def __init__(self, Name, ID, Dept):
        self.Name = Name
        self.ID = ID
        self.Dept = Dept
    
    def Display_details(self):
        print("Employee ID")
        print(f"Name : {self.Name}")
        print(f"Employee ID: {self.ID}")
        print(f"Employee Department: {self.Dept}")


Adesh = Employee_ID("Adesh Thorat", 269 , "AI engineer")
Adesh.Display_details()
```

```python
#Problem No: 3

class Contacts:
    def __init__(self, Name, Number):
         self.Name = Name
         self.Number = Number
    def DisplayDetails(self):
         
         print("Saved Contact")
         print(f"Contact Name: {self.Name}")
         print(f"Contact Number: {self.Number}")

Contact = Contacts("Adesh", 9854712254)
Contact.DisplayDetails()
```

```python
#Problem No: 4
class Product:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def display_price_tag(self):
        print("Product Information:")
        print(f"Product: {self.product_name}")
        print(f"Price: ₹{self.price}")


# Sample Usage
product = Product("Headphones", 2999)
product.display_price_tag()
```

```python
#PRoblem No: 5

class Movie:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating
    
    def DisplayDetails(self):
        print("Movie Information:")
        print(f"Movie: {self.movie_name}")
        print(f"Rating: {self.rating}/5")


movie = Movie("Avengers Endgame", 4.8)
movie.DisplayDetails()
```

```python
#Problem statement: 6

class Delivery:
    def __init__(self, customer_name, address):
         self.customer_name = customer_name
         self.address = address
    
    def DisplayDetails(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

delivery = Delivery("Adesh", "Pune")
delivery.DisplayDetails()
```
