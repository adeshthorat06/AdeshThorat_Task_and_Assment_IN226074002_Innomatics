```python
def Post_analyzer(likes):
    total_likes = 0
    # Add the likes from the list
    for like in likes:
        total_likes += like

    print(f"Total likes on the post: {total_likes}")
    #Classify the likes on post engagement
    if total_likes >= 1000:
        print("Post is Viral")
    else:
        print("Post got Normal engagement")

likes = [250,300,320,400,480]

Post_analyzer(likes)
```

```python
#Statement : 3

def Rainfall_analysis(rainfall, required_rainfall):
    total = 0
    for rain in rainfall:
        total += rain
    average = total / len(rainfall)
    print(f"The average rainfall is : {average}")
    if average < required_rainfall:
        print("Inadequate Rainfall")

    elif average >= required_rainfall:
        print("Adequate Rainfall")

rainfall = [72,60,82,95,86]
required = int(input("Enter the required Rainfall:"))
print(f"required rainfall: {required}")
Rainfall_analysis(rainfall, required)
```

```python
#Problem No: 4
def duplicate_acc(usernames):
    unique_users = set(usernames)
    # IF the length are different then there was duplicate as set removes duplicates 
    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


accounts = ["adi123", "raju45", "Aarya54", "adi123","raju45"]
duplicate_acc(accounts)
```

```python
#Problem Statement: 6

def Premium_analyzer(crops):
    premium_crops = []
    for crop in crops:
        if crop > 2000:
            premium_crops.append(crop)
    print(f"Premium Crops: {premium_crops}")

crops = [1500,1750,2000,2100,2600,7500]
Premium_analyzer(crops)
```

```python
#Problem Statement : 7
def Error_analyzer(Errors):
    total = 0
    for error in Errors:
        if error == "Error":
            total += 1
    if total == 0:
        print("System Status: Health")
    elif total <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Errors")

Sys_list = ["Error", "OK","Error","OK","Error","Error"]

Error_analyzer(Sys_list)
```

```python
#Problem No: 9
def attendance_checker(attendance_list):
    total_class = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == "P":
            attended += 1

    percentage = (attended / total_class) * 100

    print(f"Attendance Percentage: {percentage:.2f}")
    
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example Usage
attendance = ["P", "P", "A", "P", "P", "P","P","P", "A"]
attendance_checker(attendance)
```

```python
#Problem No: 10

def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        status = "Low Usage"
    elif 500 <= bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status

units = 170
bill_amount, usage_status = electricity_bill(units)

print(f"Units Consumed: {units}")
print(f"Total Bill: ₹{bill_amount}")
print(f"Usage Level: {usage_status}")
```
