```python
#Problem No: 1

def Recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount < 50:
        print("Invalid Amount: Less than 50")
        return False

    elif amount not in valid_plans :
        print("Invalid Plan select")
        print("Available plans are : 199,299,399,599")

        return False
    return True

amount = int(input("Enter the amount for recharge: "))

while not Recharge(amount):
    print("Please try Again")
    amount = int(input("Enter the amount for recharge: "))


print("Recharge Succesfully")
```

```python
#Problem No:2

def Restock(Stock ):
    for product, quantity in Stock.items():
        if quantity < 15:
            print(f"ALERT! Stock is below 15 for {product}, Please ReStock ,Available Stock: {quantity}")

        else:
            print(f"STOCK Ok for {product}")

Stock = {
"Laptops": 23,
    "Desktops": 14,
    "Monitors": 24,
    "Keyboards": 5,
    "Mice": 20
}

Restock(Stock)
```

```python
#Problem No: 3

def Result_eval(marks):
    total = 0
    for mark in marks:
        total += mark
    average = total/len(marks) 

    if average >= 50:
        return (f"Student is Pass. Average: {average}")
    else:
        return (f"Student is Fail.")

marks = [68, 42,87,92,75]
marks2 = [45,25,60,50,32]

print(Result_eval(marks))
print(Result_eval(marks2))
```

```python
#Problem No: 4
# Problem No: 4

def calc_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (per_km_rate * distance)

    if peak.lower() == "y":
        fare = fare + (fare * 0.25)

    return fare


while True:
    distance = float(input("Enter distance in KM: "))
    peak = input("Is it peak hour? (Y/N): ")

    total_fare = calc_fare(distance, peak)

    print(f"Estimated Fare: ₹{total_fare:.2f}")

    retry = input("Do you want to calc again? (Y/N): ")

    if retry.lower() != "y":
        break
```

```python
#problem No 5

def Attendance(attendance):
    total = 0
    count = 0

    for attend in attendance:
        total += 1
        if attend == 'P':
            count += 1

    percentage = int(count/total * 100)

    if percentage >= 75:
        return f"Eligible with Attendance of {percentage}%"
    else:
        return f"Not Eligible, Attendnace: {percentage}%"


attendance_list = ['P', 'P','P','A','P','P','P']

print(Attendance(attendance_list))
```

```python
#Problem 6
def Password_checker(password):
    has_digits = False
    has_special = False
    if len(password) < 8:
        return "Weak Password: Minimum 8 letters required"
    for ch in password:
        if ch.isdigit():
            has_digits = True
        if ch in "@#$":
            has_special = True
    
    if has_digits and has_special:
        return "Password is Strong"
    else: 
        return "password not Strong, include atleast One digit or Special Character @,#,$"
     

password = "Password#123"
passcode = "Hereispassword"
print(Password_checker(password))
print(Password_checker(passcode))
```
