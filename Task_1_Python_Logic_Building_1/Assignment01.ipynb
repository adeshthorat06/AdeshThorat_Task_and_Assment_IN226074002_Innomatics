```python
correct_username = "admin" #Valid Username stored in here
correct_password = "1234" #Valid password for corresponding username

username = input("Enter your Username:")
password = input("Enter your Password:")

if username == correct_username and password == correct_password:
    print("Login Successfully")
else: 
    print("Invalid Username or Password")
```

```python
marks = [45, 78, 90, 33, 60]
count_pass = 0
count_fail = 0
for mark in marks:
    if mark < 50:
        count_fail = count_fail + 1 #if Fail increment the counter by 1
    else:
        count_pass = count_pass + 1 #if Pass Increment this by 1

print(f"Number of Failed Students: {count_fail}")
print(f"Number of Pass Students: {count_pass}")
```

```python
names = [" Alice ", "bob", " CHARLIE ","Adesh   ", "  Alex "]
updated_list = [] #empty

for name in names:
    update_names = name.strip().lower() # removes spaces and lowercases the String
    updated_list.append(update_names)

print("The cleaned and standardized names:" , updated_list)
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0

for log in logs:
    if log == "ERROR": #if the given log string matches to the ERROR keyword
        error_count += 1

print(f"Total ERROR entries: {error_count}")
```
