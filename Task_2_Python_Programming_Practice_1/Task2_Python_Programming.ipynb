```python
#Problem No: 01
sentence = input("Enter the Sentence: ")
words = sentence.split() #split sentence to words

unique_list = set(words) # convert to set as it only contains unique elements

print("Unique words are :", unique_list)
print("Number of Unique words :",len(unique_list))
```

```python
#Problem No: 02
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_emp = max(employees, key=employees.get) # compare the salary and find the max one

highest_salary = employees[highest_emp]
print("highest salary employee is :", highest_emp ,":", highest_salary)
```

```python
# Problem No: 03

numbers = [45, 22, 89, 10, 66]
# We store number into the max and minimum from index 0, and if there is max or mini element we compare by loop and store it!
maximum = numbers[0] 
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("Max element: ", maximum)
print("Minimum element: ", minimum)
```

```python
#Problem No: 04

prices = [450, 1200, 899, 1500, 300, 2000, 560, 3400, 5400]
count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Price above 1000 is:", count)
```

```python
#Problem No: 5

attendance = ["P", "P", "A", "P", "P","A", "P","P"]

present = 0

for p in attendance:
    if p == "P":
        present+= 1

percentage = present/len(attendance) * 100

print("Attendance Percentage:" , percentage)
```

```python
#Problem No: 7

word = "python"

frequency = {}
for char in word:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print(frequency)
```
